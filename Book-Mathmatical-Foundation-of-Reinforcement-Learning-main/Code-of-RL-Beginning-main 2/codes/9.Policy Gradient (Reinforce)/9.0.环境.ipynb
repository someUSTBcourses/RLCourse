```python
# 定义函数
def sigmoid_c(x, c):
    return np.exp(x) / (np.exp(x) + c)

# 定义 x 范围
x = np.linspace(-10, 10, 400)

# 计算不同 c 值下的函数值
y_c0 = sigmoid_c(x, 0)
y_c1 = sigmoid_c(x, 1)
y_c2 = sigmoid_c(x, 2)
y_c5 = sigmoid_c(x, 5)
y_c10 = sigmoid_c(x, 10)
y_c100 = sigmoid_c(x, 100)

# 画图
plt.figure(figsize=(10, 8))
plt.plot(x, y_c0, label='$\\frac{e^x}{e^x + 0}$')
plt.plot(x, y_c1, label='$\\frac{e^x}{e^x + 1}$')
plt.plot(x, y_c2, label='$\\frac{e^x}{e^x + 2}$')
plt.plot(x, y_c5, label='$\\frac{e^x}{e^x + 5}$')
plt.plot(x, y_c10, label='$\\frac{e^x}{e^x + 10}$')
plt.plot(x, y_c100, label='$\\frac{e^x}{e^x + 100}$')
plt.title('Graph of $\\frac{e^x}{e^x + c}$ for different values of $c$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

# 示例数据
data = np.array([
    [0, 0, 0],
    [4, 5, 6],
    [-7, -8, -9],
    [10, 0, -10]  # 包含负数的行
])

# 使用每行的最大绝对值进行标准化
max_abs_vals = np.max(np.abs(data), axis=1, keepdims=True)
normalized_data = np.where(max_abs_vals > 0, data / max_abs_vals, 0)
normalized_data
```
